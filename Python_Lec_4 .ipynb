#case constant can be of any type



```python
x = eval(input("Enter a number "))

match x:

  case 1.1:

    print("One")

  case [1,2,3]:

    print("Two")

  case True:

    print("Three")

  case 3+4j:

    print("Four")

print("Outside match")
```

#Default Matching



```python
x = int(input("Enter a number "))

match x:

  case 1:

    print("One")

  case 2:

    print("Two")

  case 3:

    print("Three")

  case 4:

    print("Four")

  case _:
    print("Default case")
```

#More about match

1. match and case are soft keywords

2. They are not reserved words in other grammatical contexts

3. They are recognized as keywords when part of a match statement or case block only, and are allow to be used in all other contexts as variables or argument names.

4. we can use these with out strich but not a good pratice.

#duplicate case

### duplicate case is not an error but first match will work. only first case will work if correct.

```python
x = eval(input("Enter a number "))

match x:

  case 1.1:

    print("One")

  case 1.1:

    print("Two")

  case 3:

    print("Three")

  case 4:

    print("Four")

  case _:
    print("Default case")
```

# conditional casing

```python
x = eval(input("Enter a number "))

match x:

  case x if type(x)==int:

    print("One")

  case x if type(x)==float:

    print("Two")

  case x if type(x)==complex:

    print("Three")

  case _:
    print("other type")

print("Outside match")
```
