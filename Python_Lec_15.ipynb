# map ( )
The map() function in Python applies a given function to each item in an iterable (like a list) and returns a map object (iterator).


## Syntax:

map ( function , iterable ) # passed two arguemnts

```python
def square (a):

  return a*a

x = map(square , [1,2,3,4,5])

#x is a map object store id of map object

print(x)

for e in x:
   print(e, end= ' ')

l1= list(x)

print(l1)
```

## map fun ( ) with lambda

```python
numbers = [1, 2, 3]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # Output: [1, 4, 9]
```

## Example (Real Use Case):

Converting a list of prices from dollars to rupees:


```python
prices = [10, 20, 30]

dollars_to_rupees = 83  # Example rate

converted = list(map(lambda x: x * dollars_to_rupees, prices))

print(converted)  # Output: [830, 1660, 2490]
```

# filter ( )

filter (function , iterable )

1. The filter function is used to generate an output list of values that return true when the function is called.

2. used when we want to take particular values in a condition based process.


The filter() function in Python is used to filter out elements from an iterable based on a condition. It only returns the elements that satisfy the condition (i.e., return True).

## example of filter ()

```python

def f(x):

  if x%2==0:
     return x

t=(10,2,0,-4,5,8,19,9)

y=filter(f,t)

print(y)
for e in y:
   print(e, end=' ')
```

# filter with lambda expression

```python
numbers = [1, 2, 3, 4]

even = list(filter(lambda x: x%2==0, numbers))

print(even)  # Output: [2, 4]
```

# reduce function

reduce (function , iterables )

1. The reduce function applies a provided function to iterables and return a single value.

2. when we use reduce we import reduce

       from functools import reduce.

```python
from functools import reduce

x = reduce(lambda a,b: a+b, [1,2,3,4])

# a,b are input result is adding one by one 1+2 = 3 , 3+3 = 6+4 =10

print(x)
```

```python
from functools import reduce
def add(a,b):

   return a+b

l1= [1,2,3,4,5,6,7]

x= reduce(add, l1)

print("The sum of vlaues = ", x)
```
