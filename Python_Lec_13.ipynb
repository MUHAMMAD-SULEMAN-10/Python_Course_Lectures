# Recusrsion using lambda

### Claculte factorial of a number using recursive lambda expression.

```python
f = lambda  a: 1 if a==0 else a*f(a-1) # here recursively call f

print(f(5))

# f refer to a lambda fun object and store the id of this fun object.
```
