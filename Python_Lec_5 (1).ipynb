# Write  a program to print Pytho Course five times.

```python
i = 1 #initialization
while i<=5: #condition

  print("Python Course")

  i+=1 #flow
```

# write a program to print first 10 natural numbers but take input from user.

```python
num = int(input("Enter a first natural number "))

while num<=10:

  print(f"The number is {num}")

  num+=1
```

# write a program to print first 10 natural numbers in reverse order but take input from user.

```python
num = int(input("Enter natural number "))

while num>=1:

  print(f"The number is {num}")

  num-=1
```

# Write a program to print first n even natural numbers.

```python
i = 2
num = int(input("Enter a number "))

while i<=2 * num:

  print(f"The number is {i}")

  i = i+2
```

Usage:

In Python, break allows you to exit a loop when an external condition is met.

# Continue

1. continue is a keyword.

2. It can only be used in the body of loop.

3. continue transfers the control immedeately to the next iteration/repetion.  

# Example program using continue keyword.

```python
i = 1
while i <= 10:
    x = int(input("Enter a number "))
    if x % 2 == 0:
        i += 1
        continue
    print(i, "x =", x)
    i += 1
```

# Write a program to count 'a' in a given string using for loop?

```python
s = str(input("Enter a String ")) #learning platform
count = 0
for e in s:

  if e=='a':

    count+=1

print("Total occurance of 'a' is:", count)


```

# Print all the character of a string, but stop printing if 'r' appread in the sequence if all the character successfully printed the print message "All the characters are processed".

```python
x = input("Enter a String ")

for i in x:
  if i =='r':
    break
    print(i, end= '')

else:
  print("All the characters are processed")
```
